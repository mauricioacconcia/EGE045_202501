# Aprendizado Supervisionado



O aprendizado supervisionado é uma das principais abordagens de aprendizado de máquina.  
Seu objetivo é **aprender uma função de mapeamento** entre dados de entrada e saídas conhecidas, utilizando exemplos rotulados durante o processo de treinamento.

Nesse tipo de aprendizado, o modelo recebe pares de dados `(x, y)`, onde `x` representa as variáveis de entrada (características ou atributos) e `y` é a saída esperada (rótulo).  
Após o treinamento, o modelo é capaz de **generalizar** para novos exemplos, produzindo predições para dados que não foram vistos anteriormente.

O aprendizado supervisionado é amplamente utilizado em **mineração de dados**, pois permite **prever, classificar e identificar padrões** em grandes volumes de informação.


O processo de aprendizado supervisionado envolve três etapas principais:

1. **Treinamento:** o modelo ajusta seus parâmetros com base nos exemplos rotulados.  
2. **Validação:** verifica-se o desempenho do modelo em dados diferentes do treinamento.  
3. **Teste:** avalia-se a capacidade de generalização em dados totalmente novos.

Um modelo bem ajustado deve apresentar **bom desempenho em novos dados**.  
Dois problemas comuns são:
- **Overfitting:** o modelo aprende demais os detalhes do treinamento e perde capacidade de generalizar.  
- **Underfitting:** o modelo é simples demais para capturar os padrões relevantes.

O equilíbrio entre complexidade e desempenho é essencial para um aprendizado eficaz.

Os dados utilizados em aprendizado supervisionado precisam ser **rotulados**.  
Isso significa que, para cada instância de entrada, existe um valor de saída conhecido.

Exemplo:
| Entrada (x) | Saída (y) |
|--------------|-----------|
| temperatura, umidade | consumo de energia |

A saída pode ser:
- **Categórica:** quando se deseja classificar (por exemplo, “normal” ou “falha”).  
- **Contínua:** quando se deseja prever valores numéricos (por exemplo, consumo de energia em kWh).

O conjunto de dados é geralmente dividido em:
- **Treinamento (70%)**: usado para ajustar o modelo.  
- **Teste (30%)**: usado para avaliar o desempenho.  

A **validação cruzada (cross-validation)** é uma técnica importante para reduzir o viés e melhorar a confiabilidade das métricas de desempenho.

s problemas supervisionados podem ser divididos em **classificação** e **regressão**:

- **Classificação:** a variável de saída é categórica.  
  Exemplo: prever se um e-mail é “spam” ou “não spam”.  
  Métricas comuns: acurácia, precisão, recall, F1-score e matriz de confusão.

- **Regressão:** a variável de saída é contínua.  
  Exemplo: prever o consumo elétrico em kWh.  
  Métricas comuns: erro quadrático médio (MSE), erro absoluto médio (MAE) e R².

Ambos os casos exigem a análise do tipo de dado e da métrica adequada para avaliar o modelo, garantindo coerência entre o objetivo e o método escolhido.

Alguns dos algoritmos mais utilizados para tarefas de **classificação supervisionada** são:

- **Árvore de Decisão (Decision Tree):** cria divisões hierárquicas no espaço de atributos, gerando regras interpretáveis.  
- **Random Forest:** conjunto de árvores de decisão, aumentando a robustez e reduzindo o overfitting.  
- **k-Nearest Neighbors (k-NN):** classifica uma instância com base na maioria das classes de seus vizinhos mais próximos.  
- **Regressão Logística:** modelo estatístico que estima probabilidades de pertencimento a uma classe.  
- **Máquinas de Vetores de Suporte (SVM):** define hiperplanos que separam as classes com máxima margem.

Cada algoritmo possui vantagens e limitações, devendo ser escolhido conforme o tipo e a distribuição dos dados.


Na regressão, o objetivo é prever valores numéricos contínuos.  
Os algoritmos mais comuns são:

- **Regressão Linear:** estabelece uma relação linear entre as variáveis de entrada e saída.  
- **Regressão Polinomial:** estende o modelo linear para incluir termos de ordem superior.  
- **Árvore de Regressão:** divide o espaço de atributos em regiões e estima valores médios dentro de cada região.  
- **Random Forest Regressor:** combina várias árvores de regressão para obter melhor desempenho e estabilidade.

Esses métodos são amplamente aplicados em contextos de previsão de demanda, temperatura, consumo e indicadores financeiros.

O desenvolvimento de um modelo supervisionado segue etapas sistemáticas:

1. **Coleta de dados:** obtenção e armazenamento dos dados relevantes.  
2. **Pré-processamento:** limpeza, normalização e seleção de atributos.  
3. **Treinamento:** ajuste do modelo com base nos dados rotulados.  
4. **Validação:** ajuste fino de hiperparâmetros e comparação entre modelos.  
5. **Teste:** avaliação da capacidade de generalização.  
6. **Implantação:** integração do modelo em um ambiente de produção.  
7. **Monitoramento:** acompanhamento de desempenho e atualização periódica.

Essas etapas constituem o **ciclo de vida do aprendizado de máquina**, garantindo reprodutibilidade e confiabilidade no uso de modelos em aplicações reais.


O aprendizado supervisionado possui ampla aplicação em engenharia, automação e sistemas inteligentes.  
Alguns exemplos incluem:

- **Detecção de falhas em motores elétricos** a partir de dados de vibração e corrente.  
- **Previsão de demanda energética** em sistemas industriais.  
- **Controle de qualidade automatizado**, utilizando imagens e sensores.  
- **Otimização de processos** com base em dados históricos de produção.  
- **Monitoramento de desempenho** em sistemas embarcados e dispositivos IoT.

Essas aplicações demonstram a relevância prática do aprendizado supervisionado na **tomada de decisão baseada em dados**, contribuindo para maior eficiência e confiabilidade em ambientes industriais e de engenharia.

```python
# Regressão Linear Simples (passo a passo, em uma única célula)
# Objetivo: ajustar y = a*x + b usando dados sintéticos e visualizar resultados.

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

# 1) Reprodutibilidade
np.random.seed(42)

# 2) Gerar dados sintéticos: y = 3*x + 7 + ruído
n = 120
X = np.random.uniform(0, 10, size=n)
noise = np.random.normal(0, 2, size=n)
y = 3*X + 7 + noise
df = pd.DataFrame({"x": X, "y": y})

print("Regressão Linear Simples")
print("Dados sintéticos: y = 3*x + 7 + ruído\n")

# 3) Visualização inicial
plt.figure(figsize=(6,4))
plt.scatter(df["x"], df["y"], alpha=0.7)
plt.xlabel("x")
plt.ylabel("y")
plt.title("Dispersão inicial dos dados")
plt.show()

# 4) Separação treino e teste
X_train, X_test, y_train, y_test = train_test_split(
    df[["x"]].values, df["y"].values, test_size=0.25, random_state=42
)
print(f"Tamanhos: treino={len(X_train)}, teste={len(X_test)}")

# 5) Treinamento do modelo
linreg = LinearRegression()
linreg.fit(X_train, y_train)
a = linreg.coef_[0]
b = linreg.intercept_
print(f"Coeficientes estimados: a={a:.3f}, b={b:.3f}")

# 6) Avaliação no conjunto de teste
y_pred = linreg.predict(X_test)
mae  = mean_absolute_error(y_test, y_pred)
mse  = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
r2   = r2_score(y_test, y_pred)

print("\nMétricas no teste")
print(f"MAE  = {mae:.3f}")
print(f"MSE  = {mse:.3f}")
print(f"RMSE = {rmse:.3f}")
print(f"R2   = {r2:.3f}")

# 7) Visualização da reta ajustada
x_line = np.linspace(df["x"].min(), df["x"].max(), 200).reshape(-1, 1)
y_line = linreg.predict(x_line)

plt.figure(figsize=(6,4))
plt.scatter(X_train, y_train, alpha=0.6, label="Treino")
plt.scatter(X_test, y_test, alpha=0.9, marker="x", label="Teste")
plt.plot(x_line, y_line, linewidth=2, label="Reta ajustada")
plt.xlabel("x")
plt.ylabel("y")
plt.title("Regressão Linear: dados e reta ajustada")
plt.legend()
plt.show()

# 8) Gráfico de resíduos (teste)
residuos = y_test - y_pred
plt.figure(figsize=(6,4))
plt.scatter(y_pred, residuos, alpha=0.8)
plt.axhline(0, linewidth=2)
plt.xlabel("y previsto")
plt.ylabel("resíduo")
plt.title("Resíduos vs. valores previstos (teste)")
plt.show()

# 9) Predição em novos pontos
novos_x = np.array([0, 2.5, 5, 7.5, 10]).reshape(-1, 1)
novas_pred = linreg.predict(novos_x)
tabela_pred = pd.DataFrame({"x": novos_x.flatten(), "y_prevista": novas_pred})
print("\nPredições para novos x:")
print(tabela_pred.to_string(index=False))
```

## Explicação do código – Regressão Linear Simples

O exemplo apresentado demonstra o processo completo de **ajuste de um modelo de regressão linear simples**, cujo objetivo é encontrar a relação entre uma variável de entrada `x` e uma variável de saída `y`.

### 1. Geração dos dados
Foram criados dados sintéticos com a relação `y = 3x + 7 + ruído`.  
O termo de **ruído gaussiano** simula a variabilidade natural dos dados reais, tornando o problema mais realista e representando incertezas de medição.

### 2. Exploração inicial
Um **gráfico de dispersão** é exibido para observar o padrão dos dados.  
A tendência linear entre `x` e `y` indica que um modelo de regressão linear é apropriado para descrever a relação.

### 3. Divisão do conjunto de dados
O conjunto é dividido em dois subconjuntos:

- **Treinamento (75%)**: usado para ajustar os parâmetros do modelo.  
- **Teste (25%)**: usado para avaliar o desempenho do modelo em dados não vistos.

Essa separação permite estimar a **capacidade de generalização** e evita o problema de *overfitting* (ajuste excessivo aos dados de treino).

### 4. Treinamento do modelo
O modelo é criado com `LinearRegression()` da biblioteca **scikit-learn**.  
Durante o treinamento, são estimados dois parâmetros:
- `a` (**coeficiente angular**): indica a variação média de `y` para cada unidade de `x`;  
- `b` (**intercepto**): valor estimado de `y` quando `x = 0`.

A equação resultante é `y = a*x + b`, representando a melhor reta que minimiza o erro quadrático médio entre valores reais e previstos.

### 5. Avaliação do modelo
Após o ajuste, o modelo é testado com os dados de teste. As principais métricas utilizadas são:

- **MAE (Mean Absolute Error):** erro médio absoluto;  
- **MSE (Mean Squared Error):** erro médio quadrático;  
- **RMSE (Root Mean Squared Error):** raiz do erro quadrático médio;  
- **R² (Coeficiente de Determinação):** mede a proporção da variação de `y` explicada pelo modelo (valores próximos de 1 indicam bom ajuste).

Essas métricas quantificam a precisão das previsões e o quanto o modelo explica o comportamento dos dados.

### 6. Visualização da reta ajustada
O gráfico sobrepõe a **reta estimada** aos pontos observados.  
Os pontos do **treino** e do **teste** são destacados com cores diferentes, permitindo avaliar se o modelo mantém um bom desempenho em dados novos.  
A boa aderência da reta aos dados indica que o modelo capturou corretamente o padrão linear.

### 7. Análise de resíduos
O **gráfico de resíduos** mostra as diferenças entre os valores reais e previstos.  
A dispersão aleatória dos resíduos em torno de zero indica que o modelo não apresenta viés sistemático e que a suposição de linearidade é adequada.

### 8. Predição de novos valores
Por fim, o modelo é utilizado para **prever valores de `y`** para novos valores de `x`.  
Essa é uma das principais finalidades da regressão: inferir resultados para situações ainda não observadas, com base em uma relação aprendida nos dados históricos.

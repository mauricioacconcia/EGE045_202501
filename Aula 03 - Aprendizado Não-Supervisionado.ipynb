# Aprendizado de Máquina

## Supervisionado
→ Aprende a partir de exemplos rotulados (entrada + saída correta).

- **Active Learning**  
  Modelo escolhe quais dados vale a pena rotular para melhorar rápido.  

- **Multitarefa**  
  Um único modelo aprende várias tarefas relacionadas ao mesmo tempo.  

- **Online / Incremental**  
  Aprende continuamente com novos dados em fluxo (streaming).  

- **Transfer Learning**  
  Reaproveita conhecimento de um modelo pré-treinado em outra tarefa.  

- **Meta-Learning**  
  “Aprender a aprender”, adapta-se rápido a novas tarefas com poucos dados.  

## Não supervisionado
→ Encontra padrões ocultos em dados sem rótulos.

- **Semi-supervisionado**  
  Combina poucos rótulos com muitos dados não rotulados.  

- **Auto-supervisionado**  
  Cria tarefas artificiais com os próprios dados para aprender representações.  

- **Online / Incremental**  
  Atualiza o aprendizado com dados que chegam em tempo real, sem retrain completo.  

## Por Reforço
→ Agente interage com ambiente e aprende por recompensas/punições.

- **Imitação / Inverse RL**  
  Aprende a partir de demonstrações humanas (ou descobre a função de recompensa).  

- **Meta-Learning**  
  Aplica adaptação rápida em tarefas de interação ambiente.  

## Paradigmas gerais (atravessam várias categorias)

- **Federated Learning**  
  Treina de forma distribuída em vários dispositivos sem centralizar os dados, preservando privacidade.

# Aprendizado Não Supervisionado

## Origem e Contexto
O aprendizado não supervisionado surgiu como um dos ramos clássicos do aprendizado de máquina, paralelo ao supervisionado, desde os anos 1950–1960, quando pesquisadores como Tryon (1939, em psicometria) e MacQueen (1967, com o K-Means) começaram a propor métodos de agrupamento automático. Ele se consolidou em áreas como estatística multivariada, mineração de dados e mais tarde em aplicações de big data e inteligência artificial moderna. A motivação central é que, na prática, grande parte dos dados disponíveis não possui rótulos, e seria inviável ou caro rotular manualmente.

## Definição
No aprendizado não supervisionado, o algoritmo aprende padrões, estruturas e relações ocultas em dados não rotulados. Não existe uma saída esperada fornecida ao modelo; em vez disso, ele tenta organizar ou representar os dados de forma mais compreensível.

## Vantagens
- Dispensa rótulos: útil quando não há base de dados rotulada (cenário muito comum).  
- Exploração de padrões ocultos: revela estruturas não óbvias, permitindo insights iniciais.  
- Versatilidade: pode ser usado como etapa preparatória para algoritmos supervisionados.  
- Aplicável em grandes volumes de dados: útil em cenários de big data e dados não estruturados.  

## Desvantagens
- Interpretação subjetiva: os resultados nem sempre têm uma “verdade” clara, exigindo análise humana.  
- Dificuldade de avaliação: sem rótulos, é complexo medir qualidade do agrupamento ou da representação.  
- Sensibilidade a parâmetros: muitos algoritmos exigem ajuste fino (ex.: número de clusters no K-Means, ε no DBSCAN).  
- Escalabilidade: alguns métodos (hierárquicos, espectrais) não lidam bem com grandes bases de dados.  

## Principais Aplicações
- Segmentação de clientes em marketing e negócios.  
- Agrupamento de documentos e textos (ex.: motores de busca, análise de tópicos).  
- Detecção de fraudes e anomalias em finanças e segurança.  
- Bioinformática: classificação de genes, proteínas e expressões genéticas.  
- Sistemas de recomendação: encontrar itens semelhantes em catálogos de produtos.  
- Visão computacional: segmentação de imagens e pré-treinamento de modelos.

# Agrupamento (Clustering -> clusterização)

## Histórico e Origens
O conceito de agrupamento remonta à estatística multivariada e à psicometria, ainda na primeira metade do século XX. Um marco importante foi o trabalho de MacQueen (1967), que formalizou o algoritmo K-Means como método de partição. Com o crescimento da capacidade computacional e a explosão de dados digitais, o agrupamento se consolidou como uma das técnicas mais relevantes do aprendizado não supervisionado, sendo aplicado tanto na exploração inicial de dados quanto em sistemas inteligentes modernos.

## Definição
Agrupamento é a tarefa de organizar objetos em grupos (clusters) de forma que elementos dentro de um mesmo grupo sejam mais semelhantes entre si do que em relação a elementos de outros grupos. Não há rótulos definidos previamente: a estrutura é descoberta automaticamente pelo algoritmo.


## Aplicações
O agrupamento é amplamente utilizado como ferramenta exploratória e de apoio à tomada de decisão. Alguns exemplos relevantes para a engenharia elétrica incluem:

- **Sistemas de potência**: identificação de padrões de consumo de energia elétrica em diferentes perfis de clientes ou regiões.  
- **Manutenção preditiva**: agrupamento de sinais de sensores em motores, transformadores e máquinas rotativas para detectar anomalias ou regimes de operação.  
- **Processamento de sinais**: agrupamento de formas de onda ou espectros em análise de qualidade de energia.  
- **Redes de comunicação**: segmentação de tráfego de dados para otimização de redes inteligentes.  
- **Smart grids**: classificação de padrões de geração distribuída e consumo em tempo real.  
- **Visão computacional**: segmentação de imagens em inspeções industriais automatizadas.  

## Principais Algoritmos
Diversos algoritmos de agrupamento foram propostos ao longo das últimas décadas. Entre os mais conhecidos:

- **K-Means**  
- **K-Medoids**  
- **Fuzzy C-Means**  
- **Hierárquicos (Agglomerative, Divisive)**  
- **DBSCAN**  
- **OPTICS**  
- **Mean-Shift**  
- **Gaussian Mixture Models (GMM)**  
- **Spectral Clustering**  
- **Affinity Propagation**  
- **BIRCH**

### Top Tier (mais importantes para aprender)
Para aplicações práticas e estudos iniciais, três algoritmos são considerados fundamentais:
1. **K-Means**  
2. **DBSCAN**  
3. **Hierárquico (Agglomerative Clustering)**

# Algoritmo K-Means

## Contexto e Origem
O K-Means é um dos algoritmos de aprendizado **não supervisionado** mais antigos e populares.  
- A ideia inicial de particionar dados em grupos vem da estatística multivariada.  
- O termo *K-Means* foi introduzido por MacQueen em 1967.  
- O procedimento iterativo que usamos hoje é conhecido como **algoritmo de Lloyd** (1957), inicialmente proposto para problemas de quantização em telecomunicações.  

Ou seja, a motivação sempre foi: **como dividir automaticamente um conjunto de pontos em grupos que façam sentido sem rótulos prévios**.

## Intuição do Problema
Suponha que temos pontos no espaço (ex.: medições de sensores, pixels de uma imagem, sinais de consumo de energia). Queremos separar esses pontos em grupos de forma que:  
- Pontos de um mesmo grupo sejam **similares**.  
- Pontos de grupos diferentes sejam **distantes**.  

O K-Means faz isso criando **centróides** (pontos médios) que representam cada grupo.

###Formalmente

Dado um conjunto de pontos $X=\{x_i\}_{i=1}^n$, com $x_i \in \mathbb{R}^d$, e um número de clusters $K$, o K-Means procura particionar $X$ em $K$ grupos e encontrar centróides $\{\mu_k\}_{k=1}^K$ que minimizem a soma das distâncias quadráticas intra-cluster.

---

## Função Objetivo
Em otimização, a **função objetivo** é aquilo que queremos minimizar ou maximizar.  
No K-Means, o que se minimiza é a soma das distâncias quadráticas de cada ponto ao centróide do seu cluster:

$$
J = \sum_{i=1}^{n} \lVert x_i - \mu_{c(i)} \rVert^2
$$

- $x_i$: ponto de dado.  
- $\mu_{c(i)}$: centróide do cluster ao qual \(x_i\) pertence.  
- $J$: erro total de agrupamento (também chamado WCSS – *Within Cluster Sum of Squares*).  

A interpretação é simples: **quanto menor $J$, mais compactos ficam os clusters**.

## Inércia (WCSS)
O valor da função objetivo do K-Means recebe o nome de **inércia** ou **WCSS (Within-Cluster Sum of Squares)**.  
Ela mede a soma das distâncias quadráticas entre cada ponto e o centróide do seu cluster:

$$
\text{WCSS} = \sum_{i=1}^{n} \lVert x_i - \mu_{c(i)} \rVert^2
$$

Quanto menor a inércia, mais compactos estão os clusters. Porém, ela **sempre diminui** quando aumentamos $K$.  
Por isso, a inércia sozinha não serve para decidir o número de clusters, mas é útil como métrica de ajuste.


## Restrições
Cada ponto deve pertencer a **um e apenas um cluster**. Isso garante uma partição clara dos dados, sem sobreposição (diferente, por exemplo, do Fuzzy C-Means, que permite pertinência parcial).

Introduza variáveis de atribuição $r_{ik}$ que valem 1 se o ponto $x_i$ pertence ao cluster $k$ e 0 caso contrário, com restrição de partição única por ponto.

$$
\min_{\{r_{ik}\},\,\{\mu_k\}}
\; J = \sum_{i=1}^{n}\sum_{k=1}^{K} r_{ik}\,\lVert x_i - \mu_k \rVert_2^2
$$

$$
r_{ik}\in\{0,1\}, \quad \sum_{k=1}^{K} r_{ik}=1 \;\; \forall i
$$

Para $r_{ik}$ fixos, o melhor centróide é a média do cluster.

$$
\mu_k = \frac{1}{n_k}\sum_{i=1}^{n} r_{ik}\,x_i,
\qquad
n_k = \sum_{i=1}^{n} r_{ik}
$$

---

## Algoritmo de Lloyd (passo a passo)
O K-Means resolve o problema aproximando a solução por meio de um processo iterativo:

1. **Inicialização**  
   Escolhem-se \(K\) centróides iniciais (aleatórios ou com técnica *k-means++*).  

2. **Etapa de atribuição**  
   Cada ponto é atribuído ao cluster cujo centróide está mais próximo (pela distância euclidiana).  

3. **Etapa de atualização**  
   Cada centróide é recalculado como a média de todos os pontos atribuídos a ele.  

4. **Repetição**  
   Os passos 2 e 3 são repetidos até que os centróides parem de mudar ou a variação do custo \(J\) seja mínima.

###Exemplo

1. Atribuição
$$
r_i \leftarrow \arg\min_{k\in\{1,\dots,K\}} \lVert x_i - \mu_k \rVert_2^2
$$

2. Atualização
$$
\mu_k \leftarrow \frac{1}{n_k}\sum_{i: r_i = k} x_i
$$

Critério comum de parada
$$
\frac{|J^{(t)}-J^{(t-1)}|}{J^{(t-1)}} < \varepsilon
\quad \text{ou} \quad
\{\mu_k\} \text{ estável}
$$

## Inicialização k-means++
Reduz más soluções iniciais escolhendo centros com probabilidade proporcional ao quadrado da distância ao centro mais próximo.

$$
p(x) \propto D(x)^2,
\quad
D(x) = \min_{c \in \{\mu_1,\dots,\mu_m\}} \lVert x - c \rVert_2
$$

---

## Propriedades e Convergência
- Cada iteração **não aumenta o custo $J$**, porque:  
  - Ao atribuir pontos ao centróide mais próximo, o custo só pode diminuir.  
  - Ao atualizar os centróides pela média, obtemos a melhor posição possível dado o grupo.  
- Isso garante que o algoritmo **converge** em um número finito de passos.  
- Porém, a convergência é para um **mínimo local**, não necessariamente para a melhor solução global.  
- Por isso usamos várias inicializações para reduzir o risco de resultados ruins.

## Decomposição de variância
O K-Means está diretamente ligado ao conceito estatístico de **decomposição da variância**.  
Seja $\bar{x}$ a média global dos dados, temos:

$$
\underbrace{\sum_{i=1}^n \lVert x_i - \bar{x} \rVert^2}_{\text{TSS (variância total)}}
=
\underbrace{\sum_{k=1}^K \sum_{i:r_i=k} \lVert x_i - \mu_k \rVert^2}_{\text{WCSS (dentro dos clusters)}}
+
\underbrace{\sum_{k=1}^K n_k \lVert \mu_k - \bar{x} \rVert^2}_{\text{BCSS (entre os clusters)}}
$$

- **TSS (Total Sum of Squares)**: variância total em torno da média global.  
- **WCSS (Within-Cluster Sum of Squares)**: variância dentro de cada cluster, é o que o K-Means **minimiza**.  
- **BCSS (Between-Cluster Sum of Squares)**: variância explicada pelos clusters, aumenta à medida que os grupos ficam mais separados.

### Interpretação prática
- Quando o WCSS é pequeno, significa que os pontos de cada cluster estão **muito próximos do seu centróide** → clusters **coerentes** internamente.  
- Quando o BCSS é grande, significa que os centróides estão **bem afastados da média global** → clusters **distintos** uns dos outros.  
- Como o TSS é fixo (depende apenas dos dados), **minimizar o WCSS equivale a maximizar o BCSS**.  

Na prática, isso quer dizer que o K-Means busca **explicar a maior parte da variabilidade dos dados por meio dos clusters**.  
Essa mesma ideia aparece em métodos estatísticos clássicos, como a **ANOVA** (análise de variância).


---

## Escolha do número de clusters $K$
O valor de $K$ **não é encontrado automaticamente** pelo algoritmo. Ele deve ser definido antes da execução. Isso é crítico porque:  
- Se $K$ for muito pequeno, grupos diferentes podem se misturar.  
- Se $K$ for muito grande, clusters artificiais podem surgir.  

Métodos comuns para escolher $K$:  
- **Método do cotovelo**: avalia a inércia $J$ em função de $K$ e procura o ponto onde a queda do custo desacelera.  
- **Índice de Silhueta**: mede quão bem cada ponto está alocado em relação ao seu cluster e aos vizinhos.

Para um ponto $i$,

$$
a(i) = \text{distância média de } i \text{ ao seu cluster}
$$

$$
b(i) = \min_{k\neq r_i} \text{distância média de } i \text{ ao cluster } k
$$

$$
s(i) = \frac{b(i) - a(i)}{\max\{a(i), b(i)\}}
\in [-1,1]
$$

---

## Complexidade
Com $t$ iterações, custo aproximado
$$
\mathcal{O}(n\,K\,d\,t)
$$
A variante Mini-Batch K-Means reduz custo por iteração usando lotes de tamanho $b \ll n$.

## Pré-processamento e métrica
K-Means usa distância Euclidiana. Padronização é recomendada, por exemplo, fazer o z-scores nas variáveis.

## Hipóteses e Limitações
O K-Means funciona bem quando:  
- Os clusters são aproximadamente **esféricos** e de tamanho semelhante.  
- A variância dentro dos grupos não difere muito.  

Limitações:  
- **Sensível a outliers**: um ponto extremo pode deslocar um centróide.  
- **Não detecta formas complexas**: falha se os clusters tiverem formas alongadas ou não convexas.  
- **Necessita de \(K\)**: exige escolher o número de grupos antes, o que nem sempre é trivial.

---

## Resumo
O K-Means é rápido, simples e eficiente para muitos problemas práticos.  
Ele é usado em:
- Segmentação de clientes em engenharia de produção.  
- Agrupamento de formas de onda em análise de sinais elétricos.  
- Compressão de imagens e vetorização de padrões.  

É um ponto de partida clássico para entender algoritmos de agrupamento.

## PCA e visualização
Na prática, muitas vezes usamos o **PCA (Análise de Componentes Principais)** em conjunto com o K-Means.  
O PCA é um método de redução de dimensionalidade que encontra novos eixos ortogonais que explicam a maior parte da variância dos dados.  
Projetando os dados nos dois primeiros componentes principais, podemos visualizar clusters em um plano 2D:

- Isso **não altera os resultados do K-Means**, apenas facilita a interpretação.  
- Também pode acelerar o K-Means em bases muito grandes, reduzindo o número de dimensões de entrada.  



```python
# K-Means completo: Wine dataset (sem Iris), com seleção de K e gráficos
# Passo a passo, pronto para notebook

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.datasets import load_wine
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score, calinski_harabasz_score, davies_bouldin_score, adjusted_rand_score

# ============================================
# 1) Carregar dados e preparar o DataFrame
# ============================================
wine = load_wine()
X = wine.data
feature_names = wine.feature_names
y_true = wine.target  # usado apenas para avaliação externa (opcional)

df = pd.DataFrame(X, columns=feature_names)

# ============================================
# 2) Padronizar (z-score) por causa da métrica Euclidiana do K-Means
# ============================================
scaler = StandardScaler()
X_std = scaler.fit_transform(X)

# ============================================
# 3) Análise exploratória rápida: PCA 2D para visualizar estrutura
# ============================================
pca = PCA(n_components=2, random_state=42)
X_2d = pca.fit_transform(X_std)

plt.figure(figsize=(6,5))
plt.scatter(X_2d[:,0], X_2d[:,1], s=15)
plt.title("Wine: projeção PCA 2D (dados padronizados)")
plt.xlabel("PC1")
plt.ylabel("PC2")
plt.tight_layout()
plt.show()

# ============================================
# 4) Seleção de K: curva do cotovelo (inércia) e silhueta
# ============================================
Ks = range(2, 13)
inertias = []
silhouettes = []
calinskis = []
davies_bouldins = []

for k in Ks:
    km = KMeans(n_clusters=k, init="k-means++", n_init=10, max_iter=300, random_state=42)
    labels = km.fit_predict(X_std)
    inertias.append(km.inertia_)
    silhouettes.append(silhouette_score(X_std, labels))
    calinskis.append(calinski_harabasz_score(X_std, labels))
    davies_bouldins.append(davies_bouldin_score(X_std, labels))

# Plot cotovelo
plt.figure(figsize=(6,4))
plt.plot(list(Ks), inertias, marker="o")
plt.xticks(list(Ks))
plt.xlabel("K")
plt.ylabel("Inércia (WCSS)")
plt.title("Método do cotovelo")
plt.grid(True, linestyle=":")
plt.tight_layout()
plt.show()

# Plot silhueta
plt.figure(figsize=(6,4))
plt.plot(list(Ks), silhouettes, marker="o")
plt.xticks(list(Ks))
plt.xlabel("K")
plt.ylabel("Silhueta média")
plt.title("Silhueta por K")
plt.grid(True, linestyle=":")
plt.tight_layout()
plt.show()

# Métricas adicionais (opcional)
plt.figure(figsize=(6,4))
plt.plot(list(Ks), calinskis, marker="o")
plt.xticks(list(Ks))
plt.xlabel("K")
plt.ylabel("Calinski-Harabasz (maior é melhor)")
plt.title("Índice Calinski-Harabasz por K")
plt.grid(True, linestyle=":")
plt.tight_layout()
plt.show()

plt.figure(figsize=(6,4))
plt.plot(list(Ks), davies_bouldins, marker="o")
plt.xticks(list(Ks))
plt.xlabel("K")
plt.ylabel("Davies-Bouldin (menor é melhor)")
plt.title("Índice Davies-Bouldin por K")
plt.grid(True, linestyle=":")
plt.tight_layout()
plt.show()

# Sugestão automática de K pela silhueta
best_k_sil = Ks[int(np.argmax(silhouettes))]
print(f"Sugestão por silhueta: K = {best_k_sil:.0f} (silhueta = {max(silhouettes):.3f})")

# ============================================
# 5) Ajustar K-Means com o K escolhido e analisar resultados
#    Use o K sugerido pela silhueta, mas você pode sobrepor após ver os gráficos
# ============================================
K_FINAL = best_k_sil  # troque manualmente se decidir outro valor
kmeans = KMeans(n_clusters=K_FINAL, init="k-means++", n_init=10, max_iter=300, random_state=42)
labels = kmeans.fit_predict(X_std)

print(f"Inércia final (WCSS): {kmeans.inertia_:.2f}")
print(f"Silhueta final: {silhouette_score(X_std, labels):.3f}")
print(f"Calinski-Harabasz: {calinski_harabasz_score(X_std, labels):.1f}")
print(f"Davies-Bouldin: {davies_bouldin_score(X_std, labels):.3f}")

# Avaliação externa opcional com rótulos verdadeiros
ari = adjusted_rand_score(y_true, labels)
print(f"Adjusted Rand Index (com rótulos verdadeiros, só para referência): {ari:.3f}")

# ============================================
# 6) Visualização dos clusters em PCA 2D e centróides projetados
# ============================================
centers_std = kmeans.cluster_centers_               # centróides no espaço padronizado
centers_2d = pca.transform(centers_std)             # projetar centróides no espaço 2D do PCA

plt.figure(figsize=(6,5))
plt.scatter(X_2d[:,0], X_2d[:,1], c=labels, s=15)
plt.scatter(centers_2d[:,0], centers_2d[:,1], marker="X", s=200, edgecolors="k")
plt.title(f"K-Means no Wine: K={K_FINAL} em PCA 2D")
plt.xlabel("PC1")
plt.ylabel("PC2")
plt.tight_layout()
plt.show()

# ============================================
# 7) Tamanho dos clusters
# ============================================
(unique, counts) = np.unique(labels, return_counts=True)
cluster_sizes = pd.Series(counts, index=[f"Cluster {u}" for u in unique]).sort_index()
print(cluster_sizes)

plt.figure(figsize=(6,3.5))
plt.bar(cluster_sizes.index, cluster_sizes.values)
plt.ylabel("Quantidade")
plt.title("Tamanho dos clusters")
plt.tight_layout()
plt.show()

# ============================================
# 8) Interpretação: médias padronizadas por cluster (heatmap simples)
# ============================================
# média das features padronizadas por cluster
cluster_means = pd.DataFrame(X_std, columns=feature_names).groupby(labels).mean()

plt.figure(figsize=(10,5))
im = plt.imshow(cluster_means.values, aspect="auto")
plt.colorbar(im, fraction=0.046, pad=0.04)
plt.xticks(ticks=np.arange(len(feature_names)), labels=feature_names, rotation=90)
plt.yticks(ticks=np.arange(cluster_means.shape[0]), labels=[f"Cluster {i}" for i in range(cluster_means.shape[0])])
plt.title("Médias padronizadas por cluster")
plt.tight_layout()
plt.show()

# ============================================
# 9) Opcional: função utilitária para testar outros K rapidamente
# ============================================
def fit_kmeans_and_report(Xstd, k, pca_model=None, y_ref=None):
    km = KMeans(n_clusters=k, init="k-means++", n_init=10, max_iter=300, random_state=42)
    lab = km.fit_predict(Xstd)
    print(f"K={k} | Inércia={km.inertia_:.1f} | Silhueta={silhouette_score(Xstd, lab):.3f} | "
          f"CH={calinski_harabasz_score(Xstd, lab):.1f} | DB={davies_bouldin_score(Xstd, lab):.3f}")
    if y_ref is not None:
        print(f"    ARI (externo): {adjusted_rand_score(y_ref, lab):.3f}")
    if pca_model is not None:
        X2 = pca_model.transform(Xstd)
        c2 = pca_model.transform(km.cluster_centers_)
        plt.figure(figsize=(6,5))
        plt.scatter(X2[:,0], X2[:,1], c=lab, s=15)
        plt.scatter(c2[:,0], c2[:,1], marker="X", s=200, edgecolors="k")
        plt.title(f"Visualização PCA 2D para K={k}")
        plt.xlabel("PC1")
        plt.ylabel("PC2")
        plt.tight_layout()
        plt.show()

# Exemplo de uso:
# fit_kmeans_and_report(X_std, 3, pca_model=pca, y_ref=y_true)
```

# K-Means no Wine: explicação passo a passo

1) Importes e carga dos dados  
Carregamos a base `Wine` do scikit-learn (13 atributos químicos de vinhos).  
Os rótulos `y_true` existem só para **avaliação externa** (não entram no ajuste do K-Means).

2) Padronização (z-score)  
`StandardScaler()` centraliza (média 0) e escala (desvio 1) cada atributo.  
Por quê: K-Means usa distância Euclidiana; sem padronizar, variáveis em maior escala dominam a distância.

3) PCA 2D (exploração)  
`PCA(n_components=2)` só para **visualizar** a estrutura num plano 2D.  
Não é obrigatório para o K-Means, mas ajuda a inspecionar separabilidade e possíveis agrupamentos.

4) Seleção de K (cotovelo e silhueta)  
Laço em K=2..12:  
- Ajusta `KMeans(init="k-means++", n_init=10)` para cada K.  
- Guarda **inércia (WCSS)** para o **cotovelo**: tende a cair com K; procuramos o ponto onde a queda desacelera.  
- Calcula **silhueta média**: mede coesão vs separação dos clusters; valores próximos de 1 são melhores.  
- Registra **Calinski-Harabasz** (maior melhor) e **Davies-Bouldin** (menor melhor) para diagnóstico extra.  
Ao fim, sugerimos `best_k_sil` como o K que maximiza a silhueta.

5) Ajuste final com K escolhido  
`KMeans(n_clusters=K_FINAL, init="k-means++", n_init=10, max_iter=300)`  
- `k-means++` melhora a inicialização (menos mínimos ruins).  
- `n_init=10` roda 10 inicializações e pega a melhor (menor WCSS).  
Relatamos **inércia**, **silhueta**, **CH** e **DB** para checar qualidade.  
Calculamos **ARI** com `y_true` apenas como referência (não supervisiona nada, é só para avaliar).

6) Visualização em PCA + centróides  
Projetamos os centróides ajustados para o espaço 2D do PCA e plota:  
- Pontos coloridos pelos rótulos do K-Means.  
- Marcador “X” para centróides projetados.  
Interpretação: clusters bem separados no plano 2D sugerem boa segmentação (cuidado: é uma projeção).

7) Tamanho dos clusters  
Contamos elementos por cluster (barplot).  
Procure **desequilíbrios extremos**; às vezes indicam K inadequado ou dados com densidades muito desbalanceadas.

8) Interpretação por atributo (heatmap)  
`cluster_means` calcula a **média padronizada** de cada atributo por cluster.  
Heatmap mostra, por cluster, quais variáveis estão **acima/abaixo da média** global.  
Isso ajuda a **dar significado** aos grupos (ex.: “Cluster 0 tem álcool alto e prolina baixa”).

9) Função utilitária (opcional)  
`fit_kmeans_and_report` permite testar rapidamente outros K e visualizar resultados, mantendo as mesmas métricas.  
Útil para comparar decisões de K além da silhueta (ex.: trade-offs entre vários índices e a interpretação prática).

## Decisões de projeto (por que assim?)
- **Padronização**: essencial para distância Euclidiana.  
- **k-means++ + n_init**: reduz risco de mínimos locais ruins.  
- **Cotovelo + Silhueta + CH + DB**: triangulação de critérios; evita depender de um único índice.  
- **PCA apenas para visualização**: não “resolve” separabilidade, mas ajuda a interpretar e comunicar.

## Sinais de alerta e próximos passos
- **Silhueta baixa em todos os K**: talvez K-Means não seja adequado (clusters não globulares); tente DBSCAN ou GMM.  
- **Cluster muito pequeno**: pode ser ruído ou um outlier puxando centróide; avalie remoção de outliers ou outro K.  
- **Interpretação fraca no heatmap**: reveja features; considere seleção/engenharia de atributos.

# DBSCAN:

## Motivação
O DBSCAN resolve limitações clássicas do K-Means. Não exige informar o número de clusters, detecta grupos com **formas arbitrárias** e separa **ruídos** de maneira explícita. É especialmente útil quando existem **outliers** ou quando os clusters não são esféricos.

## Ideia central
A noção de cluster é substituída pela noção de **região densa** do espaço. Pontos em regiões densas formam clusters e pontos em regiões esparsas são marcados como ruído.

## Parâmetros
Há dois parâmetros principais.

1) $ \varepsilon $  
Raio que define a vizinhança de um ponto.

2) $ \text{minPts} $  
Mínimo de pontos na vizinhança de raio $ \varepsilon $ para considerar um ponto como **núcleo**. Uma regra prática comum é $ \text{minPts} \ge d+1 $, onde $ d $ é a dimensionalidade, usando valores típicos como 4, 5, 10 dependendo do ruído esperado.

## Definições formais
Seja $ D \subset \mathbb{R}^d $ o conjunto de dados e uma métrica de distância $ \text{dist}(\cdot,\cdot) $ qualquer.

1) Vizinhança de $ p $  
$$
\mathcal{N}_\varepsilon(p) = \{\, q \in D \mid \text{dist}(p,q) \le \varepsilon \,\}
$$

2) Ponto núcleo  
$$
p \text{ é núcleo} \iff |\mathcal{N}_\varepsilon(p)| \ge \text{minPts}
$$

3) Acessibilidade por densidade direta  
Um ponto $ q $ é **diretamente acessível por densidade** a partir de $ p $ se
$$
p \text{ é núcleo} \quad \text{e} \quad q \in \mathcal{N}_\varepsilon(p)
$$

4) Acessibilidade por densidade  
$ q $ é **acessível por densidade** a partir de $ p $ se existe uma cadeia $ p = p_1, p_2, \dots, p_m = q $ tal que cada $ p_{i+1} $ é diretamente acessível por densidade a partir de $ p_i $.

5) Conectividade por densidade  
Pontos $ p $ e $ q $ são **conectados por densidade** se existe um ponto $ o $ tal que $ p $ e $ q $ são acessíveis por densidade a partir de $ o $.

6) Cluster como conjunto maximal  
Um cluster $ C \subseteq D $ satisfaz:
$$
\text{(i) } \forall p,q \in C,\; p \text{ e } q \text{ são conectados por densidade}
$$
$$
\text{(ii) maximalidade: se } p \in C \text{ e } r \text{ é acessível por densidade a partir de } p,\; \text{então } r \in C
$$

7) Tipos de pontos  
Núcleo: $ |\mathcal{N}_\varepsilon(p)| \ge \text{minPts} $  
Borda: não núcleo, mas acessível por densidade a partir de algum núcleo  
Ruído: não acessível por densidade a partir de nenhum núcleo

## Algoritmo passo a passo
1) Marcar todos os pontos como não visitados.  
2) Escolher um ponto não visitado $ p $, marcar como visitado e obter $ \mathcal{N}_\varepsilon(p) $.  
3) Se $ p $ é núcleo, iniciar um novo cluster e **expandir**: adicionar todos os pontos de $ \mathcal{N}_\varepsilon(p) $. Para cada vizinho núcleo encontrado, incorporar sua vizinhança também, repetindo até não haver mais ampliações.  
4) Se $ p $ não é núcleo e não é acessível por densidade, marcar como ruído.  
5) Repetir até que todos os pontos estejam visitados.

## Complexidade
Sem estrutura de vizinhança eficiente, o custo é $ \mathcal{O}(n^2) $ por causa das buscas de vizinhos. Usando estruturas como **k-d tree** ou **ball tree** para buscas de raio, a complexidade tende a $ \mathcal{O}(n \log n) $ em dados de baixa dimensão. Em alta dimensão, a busca de vizinhos degrada e o custo pode se aproximar de $ \mathcal{O}(n^2) $.

## Escolha de $ \varepsilon $ e $ \text{minPts} $
A qualidade do DBSCAN depende fortemente desses parâmetros.

1) Heurística do gráfico de k-distâncias  
Para cada ponto, calcule a distância ao seu $ k $-ésimo vizinho mais próximo com $ k = \text{minPts} $. Ordene essas distâncias crescentemente e observe o ponto de inflexão. Esse valor é um bom candidato a $ \varepsilon $.

2) Escolha de $ \text{minPts} $  
Valores típicos: 4 ou 5 em dados com ruído moderado. Em geral, aumente $ \text{minPts} $ quando a amostra for grande ou quando houver mais ruído.

3) Padronização e métrica  
Como a decisão de densidade usa distâncias, padronizar as variáveis com escalas muito diferentes costuma ser essencial  
$$
x^{(\text{std})}_{ij} = \frac{x_{ij} - \mu_j}{\sigma_j}
$$
A métrica padrão é Euclidiana, mas é possível usar cosseno, Manhattan, haversine em dados geográficos, desde que a busca de vizinhos suporte a métrica.

## Por que DBSCAN funciona
O algoritmo constrói clusters a partir de **cadeias de pontos núcleo**, unindo regiões onde a densidade local excede um limiar implícito dado por $ \varepsilon $ e $ \text{minPts} $. Como o critério é local, o DBSCAN encontra estruturas **não convexas** e separa **ruído** sem exigir $ K $.

## Vantagens
Não requer o número de clusters.  
Identifica ruído de forma explícita.  
Detecta clusters de formas arbitrárias.  
Determinístico para parâmetros fixos e métrica fixa.

## Limitações
Densidade variável entre clusters pode confundir um $ \varepsilon $ global.  
Em alta dimensão, as distâncias tendem a se concentrar e o critério de densidade perde contraste.  
Escolher $ \varepsilon $ e $ \text{minPts} $ pode ser delicado sem heurísticas.  
Pontos de borda próximos a dois clusters podem ficar associados a apenas um dos lados.

## Relações com variantes
OPTICS relaxa a escolha global de $ \varepsilon $ e fornece uma ordenação de acessibilidade por densidade que ajuda a lidar com densidades variadas.  
HDBSCAN estende a ideia para uma hierarquia de densidades e tende a lidar melhor com densidades heterogêneas, removendo a necessidade de $ \varepsilon $.

## Quando usar DBSCAN
Quando há ruído e outliers que você deseja identificar.  
Quando os clusters têm formas complexas.  
Quando não há um $ K $ claro a priori.  
Quando a escala dos atributos pode ser normalizada e a noção de densidade faz sentido no espaço de atributos.


```python
# DBSCAN completo: seleção de eps via gráfico de k-distâncias, ajuste, avaliação e visualizações
# Pronto para colar no notebook. Usa make_moons (não é Iris), bom para clusters não convexos.

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.datasets import make_moons
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import NearestNeighbors
from sklearn.cluster import DBSCAN
from sklearn.metrics import silhouette_score, calinski_harabasz_score, davies_bouldin_score

# ==========================================================
# 1) Dados: make_moons (clusters não convexos) + ruído
# ==========================================================
X, _ = make_moons(n_samples=1500, noise=0.05, random_state=42)
X = StandardScaler().fit_transform(X)  # padronização (z-score)

plt.figure(figsize=(6, 5))
plt.scatter(X[:, 0], X[:, 1], s=10)
plt.title("Dados (make_moons) padronizados")
plt.xlabel("x1")
plt.ylabel("x2")
plt.tight_layout()
plt.show()

# ==========================================================
# 2) Escolha de minPts e gráfico de k-distâncias
#    Regra prática: minPts ~ 4 a 10 para 2D; aqui usamos 10 (robusto a ruído)
#    eps sugerido: “joelho” do gráfico das k-distâncias
# ==========================================================
minPts = 10

# k-distâncias: distância ao k-ésimo vizinho mais próximo
nbrs = NearestNeighbors(n_neighbors=minPts, metric="euclidean").fit(X)
distances, _ = nbrs.kneighbors(X)

# A distância relevante é a do k-ésimo vizinho (coluna -1)
k_dists = np.sort(distances[:, -1])

plt.figure(figsize=(6, 4))
plt.plot(k_dists)
plt.title(f"Gráfico de k-distâncias (k = {minPts})")
plt.xlabel("Pontos ordenados")
plt.ylabel(f"Distância ao {minPts}º vizinho")
plt.tight_layout()
plt.show()

# Heurística automática: percentil alto (ex.: 90º) como ponto inicial para eps
eps_sug = float(np.percentile(k_dists, 90))
print(f"Sugestão inicial para eps (p90 das k-distâncias): {eps_sug:.3f}")

# ==========================================================
# 3) Ajuste do DBSCAN com eps sugerido
# ==========================================================
db = DBSCAN(eps=eps_sug, min_samples=minPts, metric="euclidean", n_jobs=None)
labels = db.fit_predict(X)

# Informações básicas
n_clusters = len(set(labels)) - (1 if -1 in labels else 0)
n_noise = int(np.sum(labels == -1))
print(f"Clusters encontrados: {n_clusters}")
print(f"Pontos marcados como ruído: {n_noise}")

# Métricas internas (cuidado: silhouette exige >=2 clusters)
def safe_silhouette(X, labels):
    labset = set(labels)
    if -1 in labset:
        labset.remove(-1)
    if len(labset) >= 2 and len(labset) < len(X):
        mask = labels != -1
        return silhouette_score(X[mask], labels[mask])
    return np.nan

sil = safe_silhouette(X, labels)
ch = calinski_harabasz_score(X[labels != -1], labels[labels != -1]) if np.any(labels != -1) and n_clusters >= 2 else np.nan
dbi = davies_bouldin_score(X[labels != -1], labels[labels != -1]) if np.any(labels != -1) and n_clusters >= 2 else np.nan

print(f"Silhueta (sem ruído): {sil:.3f}" if not np.isnan(sil) else "Silhueta: n/d (clusters insuficientes)")
print(f"Calinski-Harabasz: {ch:.2f}" if not np.isnan(ch) else "Calinski-Harabasz: n/d")
print(f"Davies-Bouldin: {dbi:.3f}" if not np.isnan(dbi) else "Davies-Bouldin: n/d")

# Visualização: clusters (ruído marcado com 'x')
plt.figure(figsize=(6, 5))
is_noise = labels == -1
plt.scatter(X[~is_noise, 0], X[~is_noise, 1], s=10)
plt.scatter(X[is_noise, 0], X[is_noise, 1], s=30, marker='x')
plt.title(f"DBSCAN com eps={eps_sug:.3f}, minPts={minPts}  |  Clusters={n_clusters}, Ruído={n_noise}")
plt.xlabel("x1")
plt.ylabel("x2")
plt.tight_layout()
plt.show()

# ==========================================================
# 4) Varredura em eps para diagnóstico (curva de qualidade)
#    Obs.: Para DBSCAN, variar eps é crucial; aqui testamos uma grade
# ==========================================================
eps_grid = np.linspace(k_dists[int(0.70 * len(k_dists))], k_dists[int(0.98 * len(k_dists))], 15)

records = []
for eps in eps_grid:
    model = DBSCAN(eps=float(eps), min_samples=minPts, metric="euclidean")
    lab = model.fit_predict(X)
    ncl = len(set(lab)) - (1 if -1 in lab else 0)
    nno = int(np.sum(lab == -1))
    sil_g = safe_silhouette(X, lab)
    ch_g = calinski_harabasz_score(X[lab != -1], lab[lab != -1]) if np.any(lab != -1) and ncl >= 2 else np.nan
    dbi_g = davies_bouldin_score(X[lab != -1], lab[lab != -1]) if np.any(lab != -1) and ncl >= 2 else np.nan
    records.append({"eps": float(eps), "n_clusters": ncl, "n_noise": nno,
                    "silhouette": sil_g, "calinski_harabasz": ch_g, "davies_bouldin": dbi_g})

diag = pd.DataFrame.from_records(records)
display(diag)

# Gráficos de diagnóstico: n_clusters, ruído e silhueta versus eps
plt.figure(figsize=(6, 4))
plt.plot(diag["eps"], diag["n_clusters"], marker="o")
plt.xlabel("eps")
plt.ylabel("n_clusters")
plt.title("Número de clusters vs eps")
plt.tight_layout()
plt.show()

plt.figure(figsize=(6, 4))
plt.plot(diag["eps"], diag["n_noise"], marker="o")
plt.xlabel("eps")
plt.ylabel("n_noise")
plt.title("Pontos de ruído vs eps")
plt.tight_layout()
plt.show()

plt.figure(figsize=(6, 4))
plt.plot(diag["eps"], diag["silhouette"], marker="o")
plt.xlabel("eps")
plt.ylabel("Silhueta (sem ruído)")
plt.title("Silhueta vs eps")
plt.tight_layout()
plt.show()

# ==========================================================
# 5) Ajuste final (opcional): escolha manual informada por diagnóstico
#    Exemplo: pegue o eps que entrega poucos ruídos e clusters estáveis
# ==========================================================
# Escolha manual (exemplo): maior silhueta válida
if diag["silhouette"].notna().any():
    best_idx = diag["silhouette"].idxmax()
    eps_final = float(diag.loc[best_idx, "eps"])
else:
    # fallback: usar sugestão inicial
    eps_final = eps_sug

print(f"eps_final escolhido: {eps_final:.3f}")

db_final = DBSCAN(eps=eps_final, min_samples=minPts, metric="euclidean")
labels_final = db_final.fit_predict(X)

n_clusters_final = len(set(labels_final)) - (1 if -1 in labels_final else 0)
n_noise_final = int(np.sum(labels_final == -1))
sil_final = safe_silhouette(X, labels_final)

print(f"[Final] Clusters: {n_clusters_final} | Ruído: {n_noise_final} | Silhueta: {sil_final:.3f}"
      if not np.isnan(sil_final) else f"[Final] Clusters: {n_clusters_final} | Ruído: {n_noise_final} | Silhueta: n/d")

plt.figure(figsize=(6, 5))
noise_f = labels_final == -1
plt.scatter(X[~noise_f, 0], X[~noise_f, 1], s=10)
plt.scatter(X[noise_f, 0], X[noise_f, 1], s=30, marker='x')
plt.title(f"DBSCAN FINAL  |  eps={eps_final:.3f}, minPts={minPts}  |  Clusters={n_clusters_final}, Ruído={n_noise_final}")
plt.xlabel("x1")
plt.ylabel("x2")
plt.tight_layout()
plt.show()
```

# DBSCAN no make_moons: explicação passo a passo do código

## 0. Objetivo do script
Demonstrar o fluxo completo de uso do DBSCAN: escolha informada de parâmetros, ajuste, avaliação e visualização. O conjunto `make_moons` foi escolhido porque possui clusters não convexos, um cenário típico em que o DBSCAN funciona bem.

---

## 1) Geração e padronização dos dados
- `make_moons(n_samples=1500, noise=0.05, random_state=42)` cria dois arcos semicirculares com leve ruído. Isso gera dois grupos em forma de meia-lua.
- `StandardScaler().fit_transform(X)` aplica padronização z-score. No DBSCAN a decisão de densidade depende de distâncias; se as variáveis estiverem em escalas distintas, a noção de vizinhança fica distorcida. A padronização corrige esse problema.
- O primeiro `scatter` apenas confirma visualmente que os dados padronizados permanecem com a geometria de duas meias-luas.

O que observar: a nuvem deve mostrar duas estruturas curvas com pontos relativamente homogêneos ao longo das meias-luas.

---

## 2) Definição de `minPts` e gráfico de k-distâncias
- Escolha de `minPts`: regra prática em 2D é começar entre 4 e 10. O código usa `minPts = 10` para ser mais robusto ao ruído.
- `NearestNeighbors(n_neighbors=minPts)` encontra, para cada ponto, a distância ao seu k-ésimo vizinho mais próximo, que é a coluna `-1` da matriz `distances`.
- `k_dists = np.sort(distances[:, -1])` ordena essas distâncias em ordem crescente. O **gráfico de k-distâncias** costuma apresentar um **joelho**: uma região onde a curva cresce mais rapidamente. Esse ponto é um bom candidato a `eps`, pois separa pontos em regiões densas daqueles em regiões esparsas.
- O código chuta um valor inicial de `eps` usando o percentil 90 das k-distâncias: `eps_sug = np.percentile(k_dists, 90)`. É uma heurística para iniciar a busca.

Como interpretar o gráfico: procure o joelho. À esquerda do joelho, a distância ao k-ésimo vizinho é pequena, indicando regiões densas. À direita, a distância cresce rápido, sugerindo pontos mais isolados ou de borda. O valor do joelho é um bom `eps`.

---

## 3) Ajuste inicial do DBSCAN com `eps` sugerido
- `DBSCAN(eps=eps_sug, min_samples=minPts)` ajusta o modelo.
- `labels = db.fit_predict(X)` devolve rótulos de cluster. O rótulo `-1` significa **ruído**.
- Estatísticas básicas:
  - `n_clusters`: quantidade de clusters encontrados, desconsiderando o rótulo `-1`.
  - `n_noise`: quantidade de pontos marcados como ruído.

Métricas internas:
- **Silhueta**: mede coesão e separação dos clusters. O código define `safe_silhouette` que ignora ruído ao computar a silhueta e só calcula se existir pelo menos dois clusters válidos.
- **Calinski-Harabasz**: maior é melhor. Mede razão entre dispersão entre clusters e dentro dos clusters.
- **Davies-Bouldin**: menor é melhor. Mede média de similaridade entre cada cluster e seu mais parecido.

Visualização:
- Gráfico com pontos por cluster e ruído em marcador `x`. Ajuda a verificar se os arcos foram bem separados e se a quantidade de ruído faz sentido.

O que observar: se `eps` estiver muito pequeno, haverá muitos ruídos e clusters fragmentados. Se `eps` estiver muito grande, os dois arcos podem se fundir em um único cluster e o ruído desaparecer.

---

## 4) Varredura em `eps` para diagnóstico
- `eps_grid` cria uma grade de valores de `eps` em torno da região interessante das k-distâncias. A varredura é necessária porque a heurística inicial pode não ser a melhor.
- Para cada `eps` na grade:
  - O código ajusta `DBSCAN` e coleta: número de clusters, quantidade de ruído, silhueta, Calinski-Harabasz e Davies-Bouldin.
  - Os resultados são armazenados no `DataFrame diag` e três gráficos são plotados: `n_clusters vs eps`, `n_noise vs eps` e `silhouette vs eps`.

Como interpretar os gráficos:
- `n_clusters vs eps`: tipicamente começa alto para `eps` pequenos, reduz e estabiliza em um patamar. Regiões estáveis sugerem escolhas robustas de `eps`.
- `n_noise vs eps`: tende a cair quando `eps` aumenta. Muito ruído indica `eps` pequeno demais; ruído quase zero junto com `n_clusters=1` indica `eps` grande demais.
- `silhouette vs eps`: busque picos em região onde o número de clusters faz sentido para o problema. Em dados `make_moons`, um bom `eps` costuma revelar 2 clusters com silhueta moderada.

---

## 5) Escolha final e ajuste definitivo
- O código escolhe `eps_final` como o `eps` que maximiza a silhueta válida. Em aula, vale discutir que a melhor escolha pode considerar também o número de clusters, a quantidade de ruído e o contexto do domínio.
- Ajuste final com `DBSCAN(eps=eps_final, min_samples=minPts)`.
- Relatório final: número de clusters, ruído e silhueta. Novo gráfico é gerado para conferir o resultado escolhido.

O que validar: verificar se os dois arcos aparecem em cores distintas e se a quantidade de ruído é plausível. Em `make_moons`, espera-se dois clusters e algum ruído residual.

---

## 6) Decisões de projeto que aparecem no código
- Padronização por z-score: necessária porque a noção de vizinhança depende de distâncias.
- Uso do gráfico de k-distâncias: fornece um ponto de partida para `eps`.
- Varredura em `eps`: transforma o chute inicial em uma decisão informada por métricas.
- Ignorar ruído ao calcular silhueta: o ruído não pertence a cluster algum, portanto distorce a medida se incluído.
- Métricas múltiplas: silhueta, CH e DB fornecem visões complementares. Evita decidir apenas por um número.

---

## 7) Sinais de alerta e como proceder
- Somente 1 cluster para ampla faixa de `eps`: `eps` pode estar grande demais ou os dados não têm estrutura clara. Reduzir `eps` ou considerar outro algoritmo.
- Silhueta baixa em toda a grade: pode haver densidade não uniforme ou alta sobreposição. Avaliar **OPTICS** ou **HDBSCAN**.
- Ruído excessivo para quase todos os `eps`: `minPts` pode estar alto, ou o ruído dos dados é muito elevado. Reduzir `minPts` ou avaliar se a padronização e a métrica estão adequadas.
- Alta dimensionalidade: a noção de vizinhança degrada em muitas dimensões. Considerar redução de dimensionalidade ou uma métrica apropriada ao domínio.

---

## 8) Conclusão didática
O DBSCAN decide clusters pela densidade local e separa explicitamente ruído, o que o torna adequado para formas não convexas e dados com outliers. O passo crítico é a escolha informada dos parâmetros `eps` e `minPts`. O código mostra um fluxo prático: padronizar, estimar `eps` com k-distâncias, varrer uma grade ao redor, avaliar com múltiplas métricas e validar visualmente. Esse procedimento evita decisões arbitrárias e produz resultados mais explicáveis em sala e na prática.

# Agrupamento Hierárquico Aglomerativo (AHC)

## Ideia geral
O AHC constrói uma hierarquia de agrupamentos unindo, passo a passo, os pares de clusters mais “próximos” até restar apenas um. O resultado é um **dendrograma**, que mostra as fusões ao longo de diferentes níveis de similaridade. Para obter uma partição final com $K$ grupos, corta-se o dendrograma em uma **altura** apropriada.

## Distâncias entre clusters (critério de ligação)
Sejam $A$ e $B$ dois clusters, e $\text{dist}(x,y)$ uma métrica (ex.: Euclidiana).

- **Single linkage**  
  $$
  d_{\text{single}}(A,B) \;=\; \min_{x\in A,\,y\in B}\, \text{dist}(x,y)
  $$
  Favorece cadeias; pode “colar” clusters por pontes de pontos próximos.

- **Complete linkage**  
  $$
  d_{\text{complete}}(A,B) \;=\; \max_{x\in A,\,y\in B}\, \text{dist}(x,y)
  $$
  Produz grupos mais compactos; menos suscetível a pontes.

- **Average linkage (UPGMA)**  
  $$
  d_{\text{avg}}(A,B) \;=\; \frac{1}{|A|\,|B|}\sum_{x\in A}\sum_{y\in B}\text{dist}(x,y)
  $$
  Compromisso entre single e complete.

- **Ward**  
  Em vez de uma distância direta, escolhe o par cuja fusão **minimiza o aumento** da soma dos quadrados intra-cluster (variância interna). Para clusters $A$ e $B$ com tamanhos $n_A,n_B$ e centróides $\mu_A,\mu_B$, o aumento no custo é
  $$
  \Delta J_{\text{Ward}}(A,B) \;=\; \frac{n_A\,n_B}{n_A+n_B}\,\lVert \mu_A - \mu_B \rVert_2^2
  $$
  O Ward é o análogo hierárquico do K-Means: busca fusões que mantêm baixa variância dentro dos clusters.

Observação: o critério **Ward** requer a métrica Euclidiana e trabalha efetivamente com distâncias ao quadrado.

## Dendrograma e corte
O dendrograma plota, no eixo vertical, a “altura” de cada fusão, que representa a dissimilaridade no momento da união. Um **corte horizontal** separa a árvore em $K$ grupos. Alturas maiores indicam fusões entre conjuntos mais distintos.

## Cophenetic distance e qualidade do dendrograma
A distância cophenética entre dois pontos é a altura do dendrograma no nó em que eles se juntam pela primeira vez. O **coeficiente de correlação cophenética** mede quanto a árvore preserva as distâncias originais:
$$
\text{coph\_corr} \;=\; \text{corr}\big(\,\text{pdist}(X),\ \text{cophenetic}(X)\,\big)
$$
Valores mais altos indicam melhor fidelidade da hierarquia às distâncias.

## Relação com variância (Ward) e K-Means
Se $\bar{x}$ é a média global, a variância total se decompõe como
$$
\underbrace{\sum_{i=1}^n \lVert x_i - \bar{x} \rVert^2}_{\text{TSS}} \;=\;
\underbrace{\sum_{k=1}^K \sum_{i:r_i=k} \lVert x_i - \mu_k \rVert^2}_{\text{WCSS}}
\;+\;
\underbrace{\sum_{k=1}^K n_k \lVert \mu_k - \bar{x} \rVert^2}_{\text{BCSS}}
$$
O Ward escolhe fusões que pouco aumentam o WCSS, de modo que, para um dado $K$, tende a produzir partições semelhantes às do K-Means, porém sem precisar fixar $K$ previamente.

## Avaliação e escolha de $K$
- Cortar o dendrograma onde surgem **grandes saltos** de altura.  
- Medidas internas por $K$: **silhueta**, **Calinski–Harabasz**, **Davies–Bouldin**.  
- Coeficiente **cophenético** como diagnóstico da qualidade da hierarquia.

## Complexidade e pré-processamento
- Memória: $ \mathcal{O}(n^2) $ para armazenar a matriz de distâncias.  
- Tempo: implementações clássicas $ \mathcal{O}(n^3) $; com heurísticas, tipicamente $ \mathcal{O}(n^2) $ para dados de baixa dimensão.  
- Padronização (z-score) é recomendada quando atributos estão em escalas distintas. Para **Ward**, usar distância Euclidiana.

## Vantagens
- Não requer definir $K$ a priori; o dendrograma orienta cortes.  
- Produz uma **hierarquia** de soluções, útil na exploração de dados.  
- Com Ward, tende a criar clusters compactos.

## Limitações
- Escalabilidade limitada por $ \mathcal{O}(n^2) $ em memória.  
- **Single** pode sofrer com efeito cadeia; **complete** pode superfragmentar; **average** é intermediário.  
- Sensível à métrica e a outliers, especialmente com distâncias.  


```python
# Agrupamento Hierárquico Aglomerativo: exemplo completo com dendrograma, seleção de K e comparação de linkages
# Dataset: Wine (scikit-learn). Subamostramos para dendrograma mais legível.

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.datasets import load_wine
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.metrics import silhouette_score, calinski_harabasz_score, davies_bouldin_score, adjusted_rand_score
from sklearn.cluster import AgglomerativeClustering

from scipy.spatial.distance import pdist
from scipy.cluster.hierarchy import linkage, dendrogram, cophenet

# 1) Dados e padronização
wine = load_wine()
X = wine.data
y_true = wine.target  # apenas para avaliação externa opcional
feature_names = wine.feature_names

scaler = StandardScaler()
X_std = scaler.fit_transform(X)

# Subamostrar para dendrograma legível (ex.: 120 pontos estratificados)
rng = np.random.RandomState(42)
idx_per_class = [np.where(y_true == c)[0] for c in np.unique(y_true)]
sample_idx = np.concatenate([rng.choice(ix, size=min(40, len(ix)), replace=False) for ix in idx_per_class])
X_sub = X_std[sample_idx]
y_sub = y_true[sample_idx]

# 2) Dendrograma com Ward (requer distâncias Euclidianas)
D = pdist(X_sub, metric="euclidean")
Z = linkage(D, method="ward")  # linkage diretamente sobre distâncias do pdist

# Cophenetic correlation (qualidade do dendrograma)
coph_corr, _ = cophenet(Z, D)
print(f"Correlação cophenética (Ward): {coph_corr:.3f}")

plt.figure(figsize=(10, 5))
dendrogram(Z, truncate_mode=None, no_labels=True, color_threshold=None)
plt.title("Dendrograma (Ward) - Wine (subamostra)")
plt.xlabel("Amostras")
plt.ylabel("Altura (dissimilaridade)")
plt.tight_layout()
plt.show()

# 3) Escolha de K por inspeção + métricas (Ward)
# Vamos testar K=2..8 e avaliar Silhueta, CH e DB
records = []
for k in range(2, 9):
    ahc = AgglomerativeClustering(n_clusters=k, linkage="ward")
    labels = ahc.fit_predict(X_std)
    sil = silhouette_score(X_std, labels)
    ch = calinski_harabasz_score(X_std, labels)
    dbi = davies_bouldin_score(X_std, labels)
    ari = adjusted_rand_score(y_true, labels)
    records.append({"K": k, "silhouette": sil, "calinski_harabasz": ch, "davies_bouldin": dbi, "ARI": ari})

scores = pd.DataFrame.from_records(records)
print(scores)

plt.figure(figsize=(6, 4))
plt.plot(scores["K"], scores["silhouette"], marker="o")
plt.xticks(scores["K"])
plt.xlabel("K")
plt.ylabel("Silhueta")
plt.title("Silhueta por K (Ward)")
plt.tight_layout()
plt.show()

# Escolher K pela melhor silhueta (poderia cruzar com CH/DB e inspeção do dendrograma)
best_k = int(scores.loc[scores["silhouette"].idxmax(), "K"])
print(f"K escolhido (silhueta): {best_k}")

# 4) Ajuste final e visualização em PCA 2D
ahc_final = AgglomerativeClustering(n_clusters=best_k, linkage="ward")
labels_final = ahc_final.fit_predict(X_std)

pca = PCA(n_components=2, random_state=42)
X_2d = pca.fit_transform(X_std)

plt.figure(figsize=(6, 5))
plt.scatter(X_2d[:, 0], X_2d[:, 1], c=labels_final, s=15)
plt.title(f"Aglomerativo (Ward) em PCA 2D — K={best_k}")
plt.xlabel("PC1")
plt.ylabel("PC2")
plt.tight_layout()
plt.show()

print(f"Silhueta final: {silhouette_score(X_std, labels_final):.3f}")
print(f"Calinski-Harabasz final: {calinski_harabasz_score(X_std, labels_final):.1f}")
print(f"Davies-Bouldin final: {davies_bouldin_score(X_std, labels_final):.3f}")
print(f"ARI (externo, opcional): {adjusted_rand_score(y_true, labels_final):.3f}")

# 5) Comparação de linkages (single, complete, average, ward) a K fixo
# Nota: Ward exige métrica euclidiana e é menos sensível a efeito cadeia que single.
k_compare = best_k
rows = []
for link in ["single", "complete", "average", "ward"]:
    model = AgglomerativeClustering(n_clusters=k_compare, linkage=link)
    lab = model.fit_predict(X_std)
    sil = silhouette_score(X_std, lab)
    ch = calinski_harabasz_score(X_std, lab)
    dbi = davies_bouldin_score(X_std, lab)
    ari = adjusted_rand_score(y_true, lab)
    rows.append({"linkage": link, "silhouette": sil, "CH": ch, "DB": dbi, "ARI": ari})
cmp_df = pd.DataFrame(rows).sort_values("silhouette", ascending=False)
print("\nComparação de linkages a K fixo:")
print(cmp_df)

# 6) Interpretação: médias padronizadas por cluster (heatmap simples)
feat_df = pd.DataFrame(X_std, columns=feature_names)
cluster_means = feat_df.groupby(labels_final).mean()

plt.figure(figsize=(10, 5))
im = plt.imshow(cluster_means.values, aspect="auto")
plt.colorbar(im, fraction=0.046, pad=0.04)
plt.xticks(np.arange(len(feature_names)), feature_names, rotation=90)
plt.yticks(np.arange(cluster_means.shape[0]), [f"Cluster {i}" for i in range(cluster_means.shape[0])])
plt.title("Médias padronizadas por cluster (AHC Ward)")
plt.tight_layout()
plt.show()
```

# AHC no Wine: explicação passo a passo do código

## 1) Dados, padronização e subamostra
Usamos `load_wine`, que tem 13 atributos químicos. Aplicamos z-score porque as distâncias Euclidianas são sensíveis à escala. Para o **dendrograma**, subamostramos cerca de 120 amostras de forma estratificada, deixando a árvore mais legível.

## 2) Dendrograma com Ward e cophenética
- `pdist(X_sub, "euclidean")` calcula todas as distâncias por pares.  
- `linkage(D, "ward")` executa a aglomeração com o critério Ward.  
- `dendrogram(Z)` plota a hierarquia. Alturas maiores indicam fusões entre grupos mais distintos.  
- `cophenet(Z, D)` retorna a correlação entre distâncias originais e as cophenéticas. Valores mais altos indicam que a árvore preserva melhor as relações de distância.

Leitura prática: procure por grandes saltos de altura ao longo do dendrograma. Cortes logo abaixo desses saltos costumam criar partições mais naturais.

## 3) Escolha de K por métricas internas
Para Ward, testamos \(K=2\) a \(K=8\) e medimos:
- **Silhueta**: mais alto indica melhor separação relativa dos clusters.  
- **Calinski–Harabasz (CH)**: maior é melhor; razão entre dispersão entre clusters e dentro dos clusters.  
- **Davies–Bouldin (DB)**: menor é melhor; mede similaridade média ao “vizinho” mais próximo de cada cluster.  
Relatamos também o **ARI** apenas como referência externa, já que os rótulos verdadeiros existem no Wine.

A decisão de \(K\) cruza o pico de silhueta com a inspeção do dendrograma. Evita-se escolher \(K\) só por um número.

## 4) Ajuste final e visualização
Ajustamos `AgglomerativeClustering(linkage="ward", n_clusters=K)`. Projetamos com `PCA(2)` apenas para visualização. Clusters bem separados no plano indicam separação plausível, mas a decisão final deve considerar as métricas e o domínio.

## 5) Comparação entre linkages
Executamos single, complete, average e ward com \(K\) fixo:
- **Single** pode sofrer com efeito cadeia, formando clusters “esticados”.  
- **Complete** tende a criar grupos compactos, mas pode fragmentar.  
- **Average** é intermediário.  
- **Ward** minimiza aumento de variância interna, geralmente robusto para dados tabulares com Euclidiana.

Comparar silhueta, CH e DB ajuda a entender qual critério combina melhor com os dados.

## 6) Interpretação por atributos
O heatmap de **médias padronizadas** por cluster revela o perfil de cada grupo (atributos acima ou abaixo da média global). Isso transforma rótulos numéricos em descrições interpretáveis para a área de domínio.

## Sinais de alerta e próximos passos
- Dendrograma sem saltos claros e silhuetas baixas em todos os \(K\): a estrutura pode ser fraca; tente outro linkage ou outro algoritmo (ex.: GMM).  
- Resultado muito sensível ao linkage: os dados podem exigir uma métrica diferente ou engenharia de atributos.  
- Muitos outliers: considere limpeza prévia, transformação robusta ou critérios que lidem melhor com ruído (DBSCAN/HDBSCAN).



# Síntese comparativa: K-Means, DBSCAN e Aglomerativo

| Aspecto | K-Means | DBSCAN | Hierárquico Aglomerativo |
|---|---|---|---|
| Ideia central | Minimiza a soma das distâncias quadráticas aos centróides | Agrupa por densidade local e marca ruído | Funde clusters passo a passo por um critério de ligação e gera dendrograma |
| Parâmetros chave | K, init k-means++, n_init, max_iter | eps, min_samples, métrica | Linkage (single, complete, average, ward); K é definido no corte do dendrograma |
| Precisa K | Sim | Não | Não na construção; define K ao cortar a árvore |
| Forma de clusters | Aproximadamente esféricos e de tamanho similar | Arbitrária, funciona bem com formas não convexas | Depende do linkage; ward favorece grupos compactos |
| Outliers e ruído | Sensível; pode deslocar centróides | Trata ruído explicitamente com rótulo −1 | Pode ser sensível a outliers conforme a métrica e o linkage |
| Escala dos atributos | Padronização recomendada (z-score) | Padronização recomendada para distâncias | Padronização recomendada; ward requer Euclidiana |
| Métrica típica | Euclidiana | Euclidiana, Manhattan, cosseno, haversine, etc. | Euclidiana comum; ward pressupõe Euclidiana |
| Seleção e diagnóstico | Inércia (WCSS), cotovelo, silhueta | Gráfico de k-distâncias para eps, varredura de eps, silhueta sem ruído | Dendrograma e saltos de altura; silhueta, Calinski–Harabasz, Davies–Bouldin; correlação cophenética |
| Complexidade aproximada | O(n·K·d·t) | O(n log n) com estruturas de vizinhança em baixa dimensão; pode chegar a O(n²) | Memória O(n²), tempo típico O(n²); dendrograma caro para n grande |
| Escalabilidade | Muito boa; mini-batch disponível | Boa em baixa dimensão; degrada em alta dimensão | Limitada por O(n²); melhor para amostras moderadas |
| Vantagens | Simples, rápido, escalável | Não precisa K, detecta ruído e formas arbitrárias | Fornece hierarquia completa, corte flexível, bom para exploração |
| Limitações | Fixa K, sensível a outliers, falha em formas não convexas | Parâmetros sensíveis; densidades variadas dificultam; alta dimensão atrapalha | Custo e memória altos; resultados variam com linkage; efeito cadeia no single |
| Casos típicos gerais | Segmentação de clientes, compressão de imagens, agrupamento rápido de vetores | Geoespacial, IoT com ruído, detecção de anomalias, formas complexas | Bioinformática, exploração inicial, taxonomias, quando o dendrograma ajuda a decidir K |
| Exemplos em Eng. Elétrica | Perfis de carga, clusters de formas de onda padronizadas, prototipagem rápida | Eventos em qualidade de energia com ruído, padrões não convexos em redes, anomalias em sensores | Estados de operação por semelhança, hierarquia de perfis de consumo e corte guiado por dendrograma |

## Escolha rápida
1. Quer velocidade, escala e grupos aproximadamente redondos: K-Means.  
2. Espera ruído e clusters tortos ou não sabe K: DBSCAN.  
3. Precisa visualizar hierarquia e decidir K por corte: Aglomerativo (ward) com dendrograma.
